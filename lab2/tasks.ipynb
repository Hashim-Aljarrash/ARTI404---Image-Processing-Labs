```python
# step 2
def sample_image(image, factor):
    height, width = image.shape[:2]

    sample_image = cv2.resize(
        image,
        (width //  factor, height //  factor),
        interpolation= cv2.INTER_NEAREST
    )

    return sample_image

def quantize_image(image,levels):
    step = 256 // levels
    quantize_image = np.floor(image / step) * step
    quantize_image = quantize_image.astype(np.uint8)

    return quantize_image

def plot_images(original, sampled, quantized):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(sampled, cmap='gray')
    plt.title('Sampled Image')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(quantized, cmap='gray')
    plt.title('Quantized Image')
    plt.axis('off')

    plt.show()
```

```python
# step 3

image_path = "images\lena_gray_256.tif"
sampling_factor = 14
quantization_levels = 9

original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if original_image is None:
    print(f"Error: Unable to load image at {image_path}")
else:
    # Sample and quantize
    sampled_image = sample_image(original_image, sampling_factor)
    quantized_image = quantize_image(original_image, quantization_levels)

    # Plot results
    plot_images(original_image, sampled_image, quantized_image)

# --------------------------------------------------
# Arithmetic Operations
# --------------------------------------------------

from PIL import Image

# Load images
img1 = Image.open('images/lena_gray_256.tif')
img2 = Image.open('images/cameraman.tif')

# Resize images
resize = (400, 400)
img1 = img1.resize(resize, Image.Resampling.LANCZOS)
img2 = img2.resize(resize, Image.Resampling.LANCZOS)

# Convert images to arrays
im1arr = np.asarray(img1)
im2arr = np.asarray(img2)

# Image addition
addition = im1arr + im2arr

# Convert result back to image and display
resultImage = Image.fromarray(addition)
resultImage.show()


# --------------------------------------------------
# Sets and Logical Operations
# --------------------------------------------------

# Load images
img3 = Image.open('images/A.png')
img4 = Image.open('images/B.png')

img3.show()
img4.show()

# Resize images
resize = (400, 400)
img3 = img3.resize(resize, Image.Resampling.LANCZOS)
img4 = img4.resize(resize, Image.Resampling.LANCZOS)

# Convert images to arrays
im3arr = np.asarray(img3)
im4arr = np.asarray(img4)

# Logical OR (Union)
union = im3arr | im4arr

# Convert result back to image and display
resultImage2 = Image.fromarray(union)
resultImage2.show()
```

```python
# task 2

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


img1 = Image.open('images/lena_gray_256.tif').convert('L')
img2 = Image.open('images/cameraman.tif').convert('L')

resize = (256, 256)
img1 = img1.resize(resize)
img2 = img2.resize(resize)

im1 = np.asarray(img1)
im2 = np.asarray(img2)
```

```python
# task 2 / part 1

subtraction = np.clip(im1 - im2, 0, 255).astype(np.uint8)

plt.imshow(subtraction, cmap='gray')
plt.title("Image Subtraction")
plt.axis('off')
plt.show()
```

```python
# task 2 / part 2
constant_add = np.clip(im1 + 175, 0, 255).astype(np.uint8)

plt.imshow(constant_add, cmap='gray')
plt.title("Image + Constant (175)")
plt.axis('off')
plt.show()
```

```python
# task 2 / part 3

difference = im1 & (~im2)

plt.imshow(difference, cmap='gray')
plt.title("Set Difference (A - B)")
plt.axis('off')
plt.show()
```

```python
# task 2 / part 4
symmetric_difference = im1 ^ im2

plt.imshow(symmetric_difference, cmap='gray')
plt.title("Symmetric Difference")
plt.axis('off')
plt.show()
```

```python
# task 2 part 5
intersection = im1 & im2

plt.imshow(intersection, cmap='gray')
plt.title("Intersection")
plt.axis('off')
plt.show()
```
